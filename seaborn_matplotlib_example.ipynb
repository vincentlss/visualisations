---

#### Now lets create a barplot using ax.bar.
1. For our bar_heights: Choose only columns from ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']. We only want to display the first movie of the dataset (avengers) hint: use iloc[]

```python
from numpy import arange
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_heights = movie[num_cols].iloc[0].values
bar_positions = arange(5) + 0.75
fig, ax = plt.subplots()
ax.bar(bar_positions, bar_heights,0.5)
plt.show()
```

#### Assign tick positions to [0.75, 1.75, 2.75, 3.75, 4.75]

```python
### Remember to rotate x axis ticks if not column labels will clash
from numpy import arange
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_heights = movie[num_cols].iloc[0].values
bar_positions = arange(5) + 0.75
fig, ax = plt.subplots()
ax.bar(bar_positions, bar_heights,0.5)
ax.set_xticks(arange(5) + 0.75)
ax.set_xticklabels(num_cols,rotation = 90)
ax.set_xlabel('review site')
ax.set_ylabel('rating')
ax.set_title('rating by sites')
plt.show()
```

#### For our fourth week, we will be looking at a dataset for car insurance. Our main goal is to better understand people who eventually buy the insurance as compared to the rest who don't

```python
#### import pandas, seaborn as sns, matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

---

#### Lets read "insure.csv" into pandas, assign it to insure, and take a look at the dataset, here is a brief description of each columns:
group_size - How many people will be covered under the policy (1, 2, 3 or 4)<br>
homeowner - Whether the customer owns a home or not (0=no, 1=yes)<br>
car_age - Age of the customer’s car<br>
risk_factor - An ordinal assessment of how risky the customer is (1, 2, 3, 4)<br>
age_oldest - Age of the oldest person in customer's group<br>
age_youngest - Age of the youngest person in customer’s group<br>
married_couple - Does the customer group contain a married couple (0=no, 1=yes)<br>
cost - cost of the quoted coverage options <br>
purchase - 1 = insurance purchased, 0 = not purchased

```python
insure = pd.read_csv("insure.csv")
insure.head(10)
```

#### Lets remove rows containing missing values

```python
#### HINT: DataFrame.dropna()
insure = insure.dropna()
```

---
#### Using seaborn.distplot(), visualise distribution of age_oldest

```python
sns.distplot(insure['age_oldest'])
plt.show()
```

---

#### Generate kernel density plot using same column, set x label as 'oldest age'

```python
sns.kdeplot(insure['age_oldest'], shade = True)
plt.xlabel("Oldest Age")
plt.show()
```

---

#### Try out all different styles and despine all axes.

```python
sns.set_style('white')
sns.kdeplot(insure['age_oldest'], shade=True)
sns.despine(left=True, bottom=True)
plt.xlabel('Age Oldest')
plt.show()
```

---

#### Next, lets use FacetGrid method to generate 2 plots on same row:
1. One for each unique value of "purchase"
2. Shaded kernel density plot of "oldest_age" column
3. 6 inches height
4. Despine all axes
5. Display plot

```python
g = sns.FacetGrid(insure, col="purchase", height=6)
g.map(sns.kdeplot, "age_oldest", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

##### Although similar are you able to describe the differences for the group of customers who purchased against those who did not?

---

#### Lets create same plots but overlay the conditions instead

```python
g = sns.FacetGrid(insure, hue="purchase", height=6)
g.map(sns.kdeplot, "age_oldest", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

---

#### Lets make it even more presentable:
1. Remove y axis ticks(not insightful)
2. Add legend

```python
g = sns.FacetGrid(insure, hue="purchase", height=6)
g.map(sns.kdeplot, "age_oldest", shade=True)
g.add_legend()
g.set(yticks=[])
sns.despine(left=True, bottom=True)
plt.show()
```

#### Plotting heatmap
Plotting heatmap allows us to look at the value of change in a certain feature against 2 variables. 

Let’s look at an example using the same flight data, for this instance we will be seeing the change in values of number of passengers taking flights against the axis of month and year.

```python
# Load dataset
flights_long = sns.load_dataset("flights")
# Pivot the dataset from long to wide format
flights = flights_long.pivot("month", "year", "passengers")
# Create a larger figure size to plot on
f, ax = plt.subplots(figsize=(12, 6))
# Create the heat map
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax, cmap='Blues')
```
