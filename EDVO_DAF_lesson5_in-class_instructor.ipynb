### Combining dataframes

#### Given that we have two separate datasets, one dataset containing only usa companies while the other containing only non-usa companies.

```python
# First we read in the dataset containing only USA companies 
f500_only_usa = pd.read_csv('f500_only_usa.csv')

# Then we read in the dataset containing only non-USA companies 
f500_no_usa = pd.read_csv('f500_no_usa.csv')
```

Lets do a `value_counts()` on the `country` column for both dataframes to find out the number of companies from each country

```python
f500_only_usa.country.value_counts()
```

```python
f500_no_usa.country.value_counts()
```

#### Lets combine both` f500_only_usa` and `f500_no_usa` into one dataframe and name it resale

```python
f500_both = pd.concat([f500_only_usa, f500_no_usa], axis = 0)
```

---

#### We would like to create a new column "profitable" base on indivudal comapny's profit change. Conditions:
1. unprofitable: below 0
2. profitable: 0 to 100
3. very_profitable: above 100


```python
### create function to be passed into the Series.apply() method

def my_fn(profit):
    if profit < 0:
        return 'unprofitable'
    elif profit < 100:
        return 'profitable'
    else:
        return 'very_profitable'

### create new column using apply function

f500_both['profitable'] = f500_both['profit_change'].apply(my_fn)
```

#### Lets take a look at the number of unprofitable, profitable,very_profitable companies using Series.value_counts() method

```python
print(f500_both['profitable'].value_counts())
```

### Merging

#### Notice that currently our `f500_both` dataframe does not have a website column
